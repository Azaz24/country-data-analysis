# Task 1: Fetch & Validate API Response
##Purpose: Fetching live data from the internet and verifying whether the data is correct or not.

```python
# import Libraries
import requests
import pandas as pd
```

##  Fetch API Data
In this task, we fetch country data from REST API and validate the response.

```python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total Countries:", len(data))
    else:
        print("Error: Failed to fetch data")

except Exception as e:
    print("Exception occurred:", e)
```

## Task 2 — Extract & Display Data
### Purpose: To convert raw JSON data into a clean table (DataFrame).


##Data Extraction
Here, we extract required fields and convert JSON into a structured DataFrame.

```python
#Extract Required Fields & Create DataFrame
countries_list = []

for country in data:
    name = country.get("name", {}).get("common", "N/A")
    population = country.get("population", 0)
    region = country.get("region", "N/A")
    area = country.get("area", 0)

    countries_list.append({
        "common_name": name,
        "population": population,
        "region": region,
        "area": area
    })

df = pd.DataFrame(countries_list)

df.head(10)
```

## Task 3: Analyze & Summarize Country Data
### Purpose:To extract insights (real-world analysis).


## Data Analysis
We analyze the dataset to find key insights like highest population region and largest country.

```python
# Extract Insights

# Region with highest population
region_population = df.groupby("region")["population"].sum()
highest_region = region_population.idxmax()

# Country with largest area
largest_country = df.loc[df["area"].idxmax()]

print("Region with Highest Population:", highest_region)
print("Country with Largest Area:", largest_country["common_name"])
```

Output:

Region with Highest Population: Asia

Country with Largest Area: Russia
